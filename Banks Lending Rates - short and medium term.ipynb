## Lending rate is the bank rate that usually meets the short- and medium-term financing needs of the private sector. 

## This rate is normally differentiated according to creditworthiness of borrowers and objectives of financing. 

## The terms and conditions attached to these rates differ by country, however, limiting their comparability. #

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('lending rates.csv')
```

```python
df.head()
```

```python
#Dropping useless columns.

df.drop(labels=['Series Name','Series Code'], axis=1, inplace=True)
```

```python
df
```

```python
#Dropping empty lines.

df.drop(labels=[27, 28, 29, 30, 31], inplace=True)
```

```python
#Dropping lines with countries with no data.

df.drop([7,8,9,10,19,20,23,25], inplace=True)
df
```

```python
#Reseting index after having excluded several lines.

df.reset_index(drop=True, inplace=True)
df
```

```python
#Defining function to transform '..' into NaN format and the numbers (that are currently str) to float and 
#also round them to 2 decimals

def drop_points(x):
    if x == '..':
        return np.nan
    else:
        return round(float(x), 2)
```

```python
#Apllying the function to all columns with numerical values (we excluded the 2 first columns).

df.loc[:, '1990 [YR1990]':'2019 [YR2019]'] = df.drop(labels=['Country Name','Country Code'], axis=1).applymap(lambda x: drop_points(x))
```

```python
#Renaming columns with the year only.

df.rename(columns={'1990 [YR1990]':'1990', 
                   '2000 [YR2000]':'2000',
                   '2010 [YR2010]':'2010', 
                   '2011 [YR2011]':'2011', 
                   '2012 [YR2012]':'2012',
                   '2013 [YR2013]':'2013',
                   '2014 [YR2014]':'2014',
                   '2015 [YR2015]':'2015',
                   '2016 [YR2016]':'2016',
                   '2017 [YR2017]':'2017',
                   '2018 [YR2018]':'2018',
                   '2019 [YR2019]':'2019'}, inplace=True)
```

```python
import matplotlib.pyplot as plt
```

```python
#Ploting the interest rates on one graphic.

fig, ax = plt.subplots(figsize=(15,10))
ax.plot(df.drop(['Country Name','Country Code'], axis=1).columns, df.drop(['Country Name','Country Code'], axis=1).iloc[0].values, label=df['Country Code'][0], marker='o')
ax.plot(df.drop(['Country Name','Country Code'], axis=1).columns, df.drop(['Country Name','Country Code'], axis=1).iloc[1].values, label=df['Country Code'][1], marker='o')
ax.plot(df.drop(['Country Name','Country Code'], axis=1).columns, df.drop(['Country Name','Country Code'], axis=1).iloc[5].values, label=df['Country Code'][5], marker='o')
ax.plot(df.drop(['Country Name','Country Code'], axis=1).columns, df.drop(['Country Name','Country Code'], axis=1).iloc[6].values, label=df['Country Code'][6], marker='o')
ax.plot(df.drop(['Country Name','Country Code'], axis=1).columns, df.drop(['Country Name','Country Code'], axis=1).iloc[10].values, label=df['Country Code'][10], marker='o')
ax.plot(df.drop(['Country Name','Country Code'], axis=1).columns, df.drop(['Country Name','Country Code'], axis=1).iloc[13].values, label=df['Country Code'][13], marker='o')
ax.plot(df.drop(['Country Name','Country Code'], axis=1).columns, df.drop(['Country Name','Country Code'], axis=1).iloc[17].values, label=df['Country Code'][17], marker='o')

plt.grid()
ax.legend()
fig.suptitle('Banks Lending Rates', fontsize=15)


plt.show()
```
